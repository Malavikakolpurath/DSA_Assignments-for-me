5. Check and handle outliers in at least 3 columns in the dataset

```python
num_columns = ["Parch","SibSp","Age"]
```

```python
for i in num_columns:
    plt.figure()
    plt.boxplot(data[i])
    plt.title(i)
```

```python
Q1 = np.percentile(data["Parch"],25,method="midpoint")
Q2 = np.percentile(data["Parch"],50,method="midpoint")
Q3 = np.percentile(data["Parch"],75,method="midpoint")

T1 = np.percentile(data["SibSp"],25,method="midpoint")
T2 = np.percentile(data["SibSp"],50,method="midpoint")
T3 = np.percentile(data["SibSp"],75,method="midpoint")

P1 = np.percentile(data["Age"],25,method="midpoint")
P2 = np.percentile(data["Age"],50,method="midpoint")
P3 = np.percentile(data["Age"],75,method="midpoint")
```

```python
print("Q1 is",Q1)
print("Q2 is",Q2)
print("Q3 is",Q3)

print("T1 is",T1)
print("T2 is",T2)
print("T3 is",T3)

print("P1 is",P1)
print("P2 is",P2)
print("P3 is",P3)
```

```python
IQR = Q3 -Q1


low_lim = Q1 - 1.5*IQR
up_lim = Q3 + 1.5*IQR
print("min is", low_lim)
print("max is", up_lim)

ITR = T3 -T1


low_lim = T1 - 1.5*ITR
up_lim = T3 + 1.5*ITR
print("min is", low_lim)
print("max is", up_lim)

IPR = P3 -P1


low_lim = P1 - 1.5*IPR
up_lim = P3 + 1.5*IPR
print("min is", low_lim)
print("max is", up_lim)
```

```python
outlier = []
for x in data['Parch']:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
outlier
```

```python
ind1 = data["Parch"]>up_lim
data.loc[ind1].index
```

```python
data.drop(data.loc[ind1].index,inplace = True)
```
