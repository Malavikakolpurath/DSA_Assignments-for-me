```python
## spliting into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
### KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn import neighbors
metric_k=[]
neighbors = np.arange(3,15)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
for k in neighbors :
    classifier =KNeighborsClassifier(n_neighbors=k,metric='euclidean')
    classifier.fit(x_train,y_train)
    y_pred=classifier.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    metric_k.append(acc)
```

```python
metric_k
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.grid()
```

```python
classifier =KNeighborsClassifier(n_neighbors= 8,metric='euclidean')
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
acc=accuracy_score(y_test,y_pred)
metric_k.append(acc)
```

```python
print('Accuracy= ',accuracy_score(y_test,y_pred))
print('Precision= ',precision_score(y_test,y_pred))
print('Recall= ',recall_score(y_test,y_pred))
print('F1 Score= ',f1_score(y_test,y_pred))
```

#  Do k-fold and stratified k-fold cross validation techniques and find the
average accuracy score of the models

**Hold out Technique**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lrmodel=LogisticRegression()
lrmodel.fit(x_train,y_train)
```

```python
score_lr=lrmodel.score(x_test,y_test)
```

```python
score_lr
```

**K Fold cross validation**

```python
from sklearn.model_selection import KFold
```

```python
k=KFold(10)
```

```python
k
```

```python
for train_index,test_index in k.split(x,y):
  print('Training Index',train_index)
  print('Test Index',test_index)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_result=cross_val_score(lrmodel,x,y,cv=k)
```

```python
cv_result
```

```python
np.mean(cv_result)
```

**Stratified K fold Cross Validation**

```python
from sklearn.model_selection import StratifiedKFold
sk=StratifiedKFold(n_splits=10)
```

```python
for train_index,test_index in sk.split(x,y):
  print('Training Index',train_index)
  print('Test Index',test_index)
```

```python
skvc_result=cross_val_score(lrmodel,x,y,cv=sk)
```

```python
skvc_result
```

```python
np.mean(skvc_result)
```
